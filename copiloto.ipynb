```python
# Instala as bibliotecas necessárias
!pip install transformers --quiet

# Importa a pipeline de Pergunta e Resposta
from transformers import pipeline

# Cria um "copiloto" baseado em um modelo pré-treinado
copiloto = pipeline("question-answering")

# Define um contexto de domínio (você pode trocar por algo técnico, jurídico, etc)
contexto = """
O Microsoft Copilot Studio é uma ferramenta low-code que permite que os usuários criem copilotos de conversação com base em IA generativa. Ele integra o Power Virtual Agents da Microsoft e pode ser usado para automatizar fluxos de trabalho, responder perguntas frequentes e auxiliar na produtividade com dados empresariais.
"""

# Função para o usuário fazer perguntas ao copiloto
def perguntar_ao_copiloto(pergunta):
    resposta = copiloto(question=pergunta, context=contexto)
    return resposta['answer']

# Teste do copiloto
pergunta_do_usuario = "O que é o Copilot Studio?"
resposta = perguntar_ao_copiloto(pergunta_do_usuario)

print("Pergunta:", pergunta_do_usuario)
print("Resposta do Copiloto:", resposta)
```
